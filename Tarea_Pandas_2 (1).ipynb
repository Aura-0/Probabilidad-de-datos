Crear un índice

```python

def rate(overall):
    if overall <= 20:
        return "Un buen país en aspecto ecológico, social y económico"
    elif  overall >= 21:
        return "No es un buen país en aspecto ecológico, social y económico"
 
```

El método .apply() es útil cuando se requiere aplicar una función al conjunto de datos.

```python
imdb_data2["Class"] = imdb_data2["Overall"].apply(rate)

imdb_data2[["city","Planet", "Profit", "People", "Overall", "Class"]]
```
