###Experimento aleatorio: Lanzar 3 monedas juntas

Vamos a describrir nuestro espacio muestral

```python
from itertools import product
```

```python
M={"A","S"}
Omega=set(product(M,repeat=3))
Omega
```

```python
len(Omega)
```

Sea $E$ el evento: caiga águila en las 3 monedas

```python
E={omega for omega in Omega if omega [0]=="A"}
E
```

```python
F={omega for omega in Omega if omega [1]=="A"}
F
```

```python
G={omega for omega in Omega if omega [2]=="A"}
G
```

El evento $A$ es la intersección de los eventos $E$, $F$ y $G$:
$$A=E\cap F∩ G $$

```python
A= E.intersection(F.intersection(G))
A
```

```python
len(A)/len(Omega)
```

Vamos a calcular la probabilidad condicional de salgan águilas en las tres monedad dado que sabemos que cae águila en el primer lanzamiento.


```python
p=len (A.intersection(E))/len (E)
p
```

```python

Omega = set(product({1,2,3,4,5,6}, repeat =3))
Omega
```
