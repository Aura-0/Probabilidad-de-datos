##LR1

```python
x =[5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x,y)
plt.show()
```

```python
#Como puedo calcularla con respecto a la media
import numpy as np
barx = np.mean(x)
bary = np.mean(y)

print(barx,bary)
```

```python
#hay un metodo para calculae la variancia
varx = np.var(x,ddof=1)#Varianza muestral
covxy = np.cov(x,y,ddof=1)#Covarianza muestral

varx,covxy
```

```python
print(varx)
```

```python
covxy=covxy[0,1]
covxy
```

```python
beta1 = covxy/varx
beta0 = bary-beta1*barx

print(beta0,beta1)
```

```python
import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x,y)
def myfunc(x):
  return slope * x + intercept
mymodel = list(map(myfunc,x))
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```

```python
slope, intercept
```

```python
r #Es una medida (coeficiente de determinacin) que indica que tan buen modelo es el ajustado
```

```python
s = myfunc(10) # prediccin
```

```python
print(s)
```

##Tarea, Ejemplo 1
Costo de estampillas a los largo de los aos

```python
#Datos que ultilizaremos
ao =[1963,1968,1971,1974,1975,1978,1981,1985,1988,1991,1995,1999, 2001,2002,2006,2007,2008,2009,2012,2013,2014]
cost =[.05,.06,.08,.10,.13,.15,.20,.22,.25,.29,.32,.33,.34,.37,.39,.41,.42,.44,.45,.46,.49]
plt.scatter(ao,cost)
plt.show
```

```python
#Calculo de la media
barao = np.mean(ao)
barcost =np.mean(cost)
print(barao,barcost)
```

```python
#Calculo de varianza y de la covarianza
varA =np.var(ao,ddof=1)#Varianza muestral
covacost = np.cov(ao,cost,ddof=1)#Covarianza muestral
varA,covacost
```

```python
print(varA)
```

```python
covacost=covacost[0,1]
covacost
```

```python
beta1 = covacost/varA
beta0 = barcost-beta1*barao
print(beta0,beta1)
```

```python
#Grafica con la pendiente
import matplotlib.pyplot as plt
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(ao,cost)
def myfunc (ao):
  return slope * ao + intercept
mymodel = list(map(myfunc,ao))
plt.scatter(ao,cost)
plt.plot(ao,mymodel)
plt.show()
```

```python
slope, intercept
```

```python
#Calculo de r, como la medida de coeficiente de determinacin
r
```

```python
#Prediccin del costo de estampillas en el ao 2016
f = myfunc(2016)
print (f)
```
